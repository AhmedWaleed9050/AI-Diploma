```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

Exercise 1: Read Total profit of all months and show it using a line plot


```python
sales = pd.read_csv("company_sales_data.csv")
```

```python
months = sales['month_number']
total_profit = sales['total_profit']

plt.figure(figsize=(10, 6))
plt.plot(months, total_profit, marker='o', linestyle='-', color='b')
plt.title('Total Profit per Month')
plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.xticks(months)
plt.grid(True)
plt.show()
```

Exercise 2: Get total profit of all months and show line plot with the following Style properties

```python
months = sales['month_number']
total_profit = sales['total_profit']

plt.figure(figsize=(12, 6))
plt.plot(months, total_profit, marker='o', linestyle='--', color='g', linewidth=2, markersize=8, markerfacecolor='red')
plt.title('Total Profit per Month', fontsize=16, fontweight='bold')
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Total Profit', fontsize=14)
plt.xticks(months, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.axhline(y=0, color='black', linewidth=0.8)
plt.axvline(x=0, color='black', linewidth=0.8)
plt.tight_layout()
plt.show()
```

Exercise 3: Read all product sales data and show it  using a multiline plot


```python
months = sales['month_number']
facecream = sales['facecream']
facewash = sales['facewash']
toothpaste = sales['toothpaste']
bathingsoap = sales['bathingsoap']
shampoo = sales['shampoo']
moisturizer = sales['moisturizer']

plt.figure(figsize=(12, 8))

plt.plot(months, facecream, marker='o', linestyle='-', label='Face Cream')
plt.plot(months, facewash, marker='s', linestyle='-', label='Face Wash')
plt.plot(months, toothpaste, marker='^', linestyle='-', label='Toothpaste')
plt.plot(months, bathingsoap, marker='D', linestyle='-', label='Bathing Soap')
plt.plot(months, shampoo, marker='*', linestyle='-', label='Shampoo')
plt.plot(months, moisturizer, marker='x', linestyle='-', label='Moisturizer')

plt.title('Sales Data for Various Products', fontsize=16, fontweight='bold')
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Sales Units', fontsize=14)

plt.legend(title='Products', fontsize=12)

plt.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.xticks(months, fontsize=12)
plt.yticks(fontsize=12)

plt.tight_layout()
plt.show()
```

Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot

```python
months = sales['month_number']
toothpaste_sales = sales['toothpaste']

plt.figure(figsize=(10, 6))
plt.scatter(months, toothpaste_sales, color='blue', marker='o')
plt.title('Toothpaste Sales per Month', fontsize=16, fontweight='bold')
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Sales Units', fontsize=14)
plt.xticks(months, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()
plt.show()
```

Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

```python
import numpy as np

months = sales['month_number']
facecream_sales = sales['facecream']
facewash_sales = sales['facewash']

bar_width = 0.35

r1 = np.arange(len(months))
r2 = [x + bar_width for x in r1]

plt.figure(figsize=(12, 8))
plt.bar(r1, facecream_sales, color='blue', width=bar_width, edgecolor='grey', label='Face Cream')
plt.bar(r2, facewash_sales, color='green', width=bar_width, edgecolor='grey', label='Face Wash')

plt.title('Face Cream and Face Wash Sales per Month', fontsize=16, fontweight='bold')
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Sales Units', fontsize=14)
plt.xticks([r + bar_width / 2 for r in range(len(months))], months)
plt.legend()

plt.grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```

Exercise 6: Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk

```python
months = sales['month_number']
bathingsoap_sales = sales['bathingsoap']

plt.figure(figsize=(12, 8))
plt.bar(months, bathingsoap_sales, color='purple', edgecolor='grey')

plt.title('Bathing Soap Sales per Month', fontsize=16, fontweight='bold')
plt.xlabel('Month Number', fontsize=14)
plt.ylabel('Sales Units', fontsize=14)
plt.xticks(months, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.savefig('bathing_soap_sales.png')
plt.show()
```

Exercise 7: Read the total profit of each month and show it using the histogram to see the most common profit ranges

```python
plt.figure(figsize=(10, 6))
plt.hist(sales['total_profit'], bins=5, color='skyblue', edgecolor='black')

plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Histogram of Total Profit')

plt.show()
```

Exercise 8: Calculate total sale data for last year for each product and show it using a Pie chart


```python
total_sales = sales.drop(columns=['month_number', 'total_units', 'total_profit']).sum()

plt.figure(figsize=(10, 8))
plt.pie(total_sales, labels=total_sales.index, autopct='%1.1f%%', startangle=140)
plt.title('Total Sales Distribution for Last Year')
plt.axis('equal')

plt.show()
```

Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

```python
bathing_soap_data = sales[['month_number', 'bathingsoap']]
facewash_data = sales[['month_number', 'facewash']]

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10, 8))

axes[0].plot(bathing_soap_data['month_number'], bathing_soap_data['bathingsoap'], marker='o', color='b')
axes[0].set_title('Bathing Soap Sales Over Months')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Units Sold')

axes[1].plot(facewash_data['month_number'], facewash_data['facewash'], marker='o', color='r')
axes[1].set_title('Facewash Sales Over Months')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Units Sold')

plt.tight_layout()

plt.show()
```

Exercise Question 10: Read all product sales data and show it using the stack plot


```python
product_data = sales.drop(columns=['month_number', 'total_units', 'total_profit'])

product_data_transposed = product_data.T

plt.figure(figsize=(10, 6))
plt.stackplot(sales['month_number'], product_data_transposed, labels=product_data_transposed.index)
plt.title('Sales Data for All Products')
plt.xlabel('Month')
plt.ylabel('Units Sold')
plt.legend(loc='upper left')

plt.show()
```
