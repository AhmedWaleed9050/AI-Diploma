### Step 1. Import the necessary libraries:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv).

### Step 3. Assign it to a variable called tips

```python
tips = pd.read_csv('tips.csv')
```

### Step 4. Delete the Unnamed 0 column

```python
tips = tips.drop(columns=['Unnamed: 0'])
```

### Step 5. Plot the total_bill column histogram

```python
plt.hist(tips['total_bill'], bins=10, edgecolor='black')
plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
plt.scatter(tips['total_bill'], tips['tip'])
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
from pandas.plotting import scatter_matrix
scatter_matrix(tips, figsize=(8, 8))
plt.suptitle('Scatter Plot Matrix of Total Bill, Tip, and Size')
plt.show()
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(8, 6))
plt.boxplot([tips[tips['day'] == 'Sun']['total_bill'],
             tips[tips['day'] == 'Sat']['total_bill'],
             tips[tips['day'] == 'Thur']['total_bill'],
             tips[tips['day'] == 'Fri']['total_bill']],
            labels=['Sun', 'Mon', 'Tue', 'Wed'])
plt.title('Box Plot of Total Bill Across Days')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
colors = {'Male': 'blue', 'Female': 'red'}

plt.figure(figsize=(8, 6))
for sex, group in tips.groupby('sex'):
    plt.scatter(group['tip'], group['day'], label=sex, color=colors[sex], alpha=0.5)

plt.title('Scatter Plot of Tip vs Day by Sex')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.legend()
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='day', y='total_bill', hue='time', data=tips)
plt.title('Total Bill per Day by Time')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Time')
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
tips_dinner = tips[tips['time'] == 'Dinner']
tips_lunch = tips[tips['time'] == 'Lunch']

# Create two histograms side by side
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(tips_dinner['tip'], bins=5, color='blue', alpha=0.7)
plt.title('Tip Distribution for Dinner')
plt.xlabel('Tip')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(tips_lunch['tip'], bins=5, color='green', alpha=0.7)
plt.title('Tip Distribution for Lunch')
plt.xlabel('Tip')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
male_data = tips[tips['sex'] == 'Male']
plt.scatter(male_data[male_data['smoker'] == 'Yes']['total_bill'], male_data[male_data['smoker'] == 'Yes']['tip'], color='blue', label='Male Smoker', alpha=0.7)
plt.scatter(male_data[male_data['smoker'] == 'No']['total_bill'], male_data[male_data['smoker'] == 'No']['tip'], color='red', label='Male Non-Smoker', alpha=0.7)
plt.title('Male: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()

plt.subplot(1, 2, 2)
female_data = tips[tips['sex'] == 'Female']
plt.scatter(female_data[female_data['smoker'] == 'Yes']['total_bill'], female_data[female_data['smoker'] == 'Yes']['tip'], color='blue', label='Female Smoker', alpha=0.7)
plt.scatter(female_data[female_data['smoker'] == 'No']['total_bill'], female_data[female_data['smoker'] == 'No']['tip'], color='red', label='Female Non-Smoker', alpha=0.7)
plt.title('Female: Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.legend()

plt.tight_layout()
plt.show()
```
